### relative grading analysis

```python
import pandas as pd
```

```python
df = pd.read_csv('NPTEL-april-2024.xlsx - Sheet1.csv')
# data from NSUT Notices
```

```python
def relative_grading(roll_no):
    score = df[df['Roll Number'] == roll_no]['Final Score'].iloc[0]
    mean = df[df['Roll Number'] == roll_no]['Course mean'].iloc[0]
    std = df[df['Roll Number'] == roll_no]['Course std'].iloc[0]
    N_score = (score - mean)/std
    if N_score > 1.5:
        return 10
    elif N_score > 1.0 and N_score <= 1.5:
        return 9
    elif N_score > 0.5 and N_score <= 1.0:
        return 8
    elif N_score > 0.0 and N_score <= 0.5:
        return 7
    elif N_score > -0.5 and N_score <= 0.0:
        return 6
    elif N_score > -1.0 and N_score <= -0.5:
        return 5
    elif N_score > -1.5 and N_score <= -1.0:
        return 4
    else:
        return 0
```

```python
def absolute_grading(roll_no):
    score = df[df['Roll Number'] == roll_no]['Final Score'].iloc[0]
    if score >= 90 and score <= 100:
        return 10
    elif score >= 81 and score < 90:
        return 9
    elif score >= 72 and score < 81:
        return 8
    elif score >= 63 and score < 72:
        return 7
    elif score >= 54 and score < 63:
        return 6
    elif score >= 45 and score < 54:
        return 5
    elif score >= 35 and score < 45:
        return 4
    else:
        return 0
```

```python
df['Final Score'] = df['Final Score'].where(df['Certificate Type'] != 'No Certificate', 0)
```

```python
df.head()
```

```python
def number_of_students(subject):
    return len(df[df['Course Name'] == subject])
```

```python
def grading_type(subject):
    if number_of_students(subject) <= 30:
        return 'absolute'
    else:
        return 'relative'
```

```python
df['Number of students'] = df['Course Name'].apply(number_of_students)
df['Grading Type'] = df['Course Name'].apply(grading_type)
```

```python
def mean(subject):
    return df[df['Course Name'] == subject]['Final Score'].mean()
```

```python
def std(subject):
    return df[df['Course Name'] == subject]['Final Score'].std()
```

```python
df['Course mean'] = df['Course Name'].apply(mean)
```

```python
df['Course std'] = df['Course Name'].apply(std)
```

```python
df.head()
```

```python
def final_grade(roll_number):
    grading_type = df[df['Roll Number'] == roll_number]['Grading Type'].iloc[0]
    if grading_type == 'absolute':
        return absolute_grading(roll_number)
    else:
        return relative_grading(roll_number)
```

```python
df.head()
```

```python
df['GPA'] = df['Roll Number'].apply(final_grade)
```
