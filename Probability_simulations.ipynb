## Simulation Tossing of a coin

```python
import matplotlib.pyplot as plt
import random

data = [0]*2
labels = ['H', 'T']
for i in range(1000000):
    data[random.randint(0,1)] += 1

data = [(x/sum(data))*100 for x in data]

fig, ax = plt.subplots()
bars = ax.bar(range(len(data)), data, color=['red', 'blue'])
ax.set_xticks([])
ax.set_xticks(range(len(data)))
ax.set_xticklabels(labels)
ax.set_xlabel('Coin side')
ax.set_ylabel('Frequency of occurance (in %)')
plt.show()
```

## Throwing a Dice simulation

```python
import matplotlib.pyplot as plt
import random

data = [0]*6
labels = ['1','2','3','4','5','6']
for i in range(1000000):
    data[random.randint(0,5)] += 1

data = [(x/sum(data))*100 for x in data]

fig, ax = plt.subplots()
bars = ax.bar(range(len(data)), data, color=['r','g','b','m','y','b'])
ax.set_xticks([])
ax.set_xticks(range(len(data)))
ax.set_xticklabels(labels)
ax.set_xlabel('Dice face')
ax.set_ylabel('Frequency of occurance (in %)')
plt.show()
```
